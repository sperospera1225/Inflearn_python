## 크롤링한 데이터에서 다시 크롤링하기

```python
import requests,openpyxl
from bs4 import BeautifulSoup

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['순위', '제품명', '가격', '링크', '제조사'])
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 20
excel_sheet.column_dimensions['D'].width = 80
excel_sheet.column_dimensions['E'].width = 20
num = 0

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('div:nth-child(5) > div:nth-child(3) > ul > li')

# ul까지 먼저 호출한다음 price와 naem을 추출함으로써 mismatch가 일어나지 않도록한다
# 이 경우 만약 price나 name이 호출되지않으면 none의값을 가지므로 get_text에서 오류가 발생함
for index, item in enumerate(data):
    price = item.select_one('div.s-price span')
    name = item.select_one('a.itemname')
    num += 1
# 크롤링한 데이터에서 다시 크롤링하기 
    res_info = requests.get(name['href']) 
# a태그안의 특정속성의 값을 가져올려면 사전구조처럼 대괄호 안에 속성명을 넣으면 됨    
    soup_info = BeautifulSoup(res_info.content, 'html.parser')
    data_info = soup_info.select_one('#container p.shoptit > a > strong')
    print(num)
    excel_sheet.append([num, name.get_text(), price.get_text(), name['href'], data_info.get_text()])
    excel_sheet.cell(row = num+1, column = 4).hyperlink = name['href']
# A1셀 데이터 가운데 정렬
cell_A1 = excel_sheet['A1'] # 셀 선택하기
cell_A1.alignment = openpyxl.styles.Alignment(horizontal='center') # 가운데 정렬
cell_A1.font = openpyxl.styles.Font(color='651FFF') # 폰트색깔바꾸기: google material color palette
#https://material.io/design/color/the-color-system.html#tools-for-picking-colors
cell_B1 = excel_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_B1.font = openpyxl.styles.Font(color='651FFF')

cell_C1 = excel_sheet['C1']
cell_C1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_C1.font = openpyxl.styles.Font(color='651FFF')

cell_D1 = excel_sheet['D1']
cell_D1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_D1.font = openpyxl.styles.Font(color='651FFF')

cell_E1 = excel_sheet['E1']
cell_E1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_E1.font = openpyxl.styles.Font(color='651FFF')

excel_file.save('gmarket.xlsx')
excel_file.close()
```
