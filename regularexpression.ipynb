```python
import openpyxl
import re

regex = re.compile(' [A-Za-z]+\.')
# 엑셀파일 열기
work_book = openpyxl.load_workbook('train.xlsx')

# 현재 Active Sheet 얻기
work_sheet = work_book.active

# work_sheet.rows는 해당 쉬트의 모든 행을 객체로 가지고 있음
for each_row in work_sheet.rows:
    print (each_row[3].value)
    print (regex.findall(each_row[3].value))    

work_book.close()
```

### 이름 데이터만 보고 성별관련 정보 알아내기

```python
import openpyxl
import re

regex = re.compile(' [A-Za-z]+\.')
# 엑셀파일 열기
work_book = openpyxl.load_workbook('01_data/train.xlsx')

# 현재 Active Sheet 얻기
work_sheet = work_book.active

# work_sheet.rows는 해당 쉬트의 모든 행을 객체로 가지고 있음
for each_row in work_sheet.rows:
    print (each_row[3].value)
    print (regex.findall(each_row[3].value))
    if len(regex.findall(each_row[3].value)) > 0:
        if regex.findall(each_row[3].value)[0].strip() == "Mr.":
            print ("남성")

work_book.close()
```

### 2. 반복 ? , \* , +
* ? 는 앞 문자가 0번 또는 1번 표시되는 패턴 (없어도 되고, 한번 있어도 되는 패턴)
* \* 는 앞 문자가 0번 또는 그 이상 반복되는 패턴
* \+ 는 앞 문자가 1번 또는 그 이상 반복되는 패턴

### ? 사용 예

```python
pattern = re.compile('D?A')     # 앞에 문자 D가 없거나, 여러번 반복되고 마지막이 A 인 문자열
```

```python
pattern.search("DDDDDDA")
```

```python
import openpyxl
import re

work_file = openpyxl.load_workbook('data_kr.xlsx')
work_sheet = work_file.active
for row in work_sheet.rows:
    pattern = re.compile('[0-9]+')
    findalled = pattern.findall(row[1].value)
    if len(findalled) == 0:
        continue
    splited = re.split('-',row[1].value)
    subed = re.sub('[0-9]{7}','*******',splited[1])
    print(splited[0]+"-"+subed)
work_file.close()
```

```python
import openpyxl
import re

# 1) train.xlsx 파일 로드하고, 각 쉬트 생성하기
work_book = openpyxl.load_workbook('train.xlsx')
work_sheet = work_book.active
work_sheet_man = work_book.create_sheet('남성')
work_sheet_solo_woman = work_book.create_sheet('미혼여성')
work_sheet_married_woman = work_book.create_sheet('기혼여성')
work_sheet_others = work_book.create_sheet('기타')
work_sheet_report = work_book.create_sheet('분석보고서')

# 2) 분석에 필요한 데이터 변수 초기화하기
man_survived, man_unsurvived = 0, 0
solo_woman_survived, solo_woman_unsurvived = 0, 0
married_woman_survived, married_woman_unsurvived = 0, 0
others_survived, others_unsurvived = 0, 0
regex = re.compile(' [A-Za-z]+\.')

# 3) train.xlsx 파일의 train 쉬트 데이터를 분석해서, 각 쉬트에 데이터 복사하기
for each_row in work_sheet.rows:
    print (each_row[3].value)
    data = regex.findall(each_row[3].value)
    
    # 3-1) train 쉬트의 첫 행은 항목 이름이 있으므로, 이를 각 쉬트에 복사하기
    if each_row[0].row == 1:
        work_sheet_man.append([each_item.value for each_item in each_row])
        work_sheet_solo_woman.append([each_item.value for each_item in each_row])
        work_sheet_married_woman.append([each_item.value for each_item in each_row])
        work_sheet_others.append([each_item.value for each_item in each_row])
    else:
        # 3-2) train 쉬트의 각 행의 이름을 분석해서, 알맞은 쉬트에 복사하기
        if len(data) > 0:
            if data[0] == ' Mr.':
                work_sheet_man.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    man_survived += 1
                else:
                    man_unsurvived += 1

            elif data[0] == ' Miss.':
                work_sheet_solo_woman.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    solo_woman_survived += 1
                else:
                    solo_woman_unsurvived += 1

            elif data[0] == ' Mrs.':
                work_sheet_married_woman.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    married_woman_survived += 1
                else:
                    married_woman_unsurvived += 1

            else:
                work_sheet_others.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    others_survived += 1
                else:
                    others_unsurvived += 1

# 4) 리포트 쉬트 만들기
work_sheet_report.column_dimensions['A'].width = 5
work_sheet_report.column_dimensions['B'].width = 10
work_sheet_report.column_dimensions['C'].width = 10
work_sheet_report.column_dimensions['D'].width = 10

work_sheet_report.append(['', '', '', '', ''])
work_sheet_report.append(['', '분류', '생존자수', '사망자수', '생존률'])

survived_rate = "%.2f%%" % (man_survived / (man_survived + man_unsurvived) * 100)
work_sheet_report.append(['', '남성', man_survived, man_unsurvived, survived_rate])

solo_woman_survived_rate = "%.2f%%" % (solo_woman_survived / (solo_woman_survived + solo_woman_unsurvived) * 100)
work_sheet_report.append(['', '미혼여성', solo_woman_survived, solo_woman_unsurvived, solo_woman_survived_rate])

married_woman_survived_rate = "%.2f%%" % (married_woman_survived / (married_woman_survived + married_woman_unsurvived) * 100)
work_sheet_report.append(['', '미혼여성', married_woman_survived, married_woman_unsurvived, married_woman_survived_rate])

others_survived_rate = "%.2f%%" % (others_survived / (others_survived + others_unsurvived) * 100)
work_sheet_report.append(['', '미혼여성', others_survived, others_unsurvived, others_survived_rate])

# 4-1) 위 값들이 정상적인지 확인하기 위해 사전에는 출력해보았음
# print (man_survived, man_unsurvived, "%.2f%%" % (man_survived / (man_survived + man_unsurvived) * 100)
# print (solo_woman_survived, solo_woman_unsurvived, "%.2f%%" % (solo_woman_survived / (solo_woman_survived + solo_woman_unsurvived) * 100))
# print (married_woman_survived, married_woman_unsurvived, "%.2f%%" % (married_woman_survived / (married_woman_survived + married_woman_unsurvived) * 100))
# print (others_survived, others_unsurvived, "%.2f%%" % (others_survived / (others_survived + others_unsurvived) * 100))

# 5) 각 데이터 저장하기
work_book.save('answer.xlsx')
work_book.close()
```
